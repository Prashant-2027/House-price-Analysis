# Data Visualization

```python
df.hist(layout=(4,4),figsize=(20,20))
```

```python
from sklearn.utils import resample
df_majority = df[(df['CHAS']==0)]
df_minority = df[(df['CHAS']==1)]
df_minority_upsampled = resample(df_minority,
                                 replace = True,
                                 n_samples = 350,
                                 random_state = 42)
df = pd.concat([df_minority_upsampled, df_majority])
df.head()
```

```python
from sklearn.utils import resample
df_majority = df[(df['ZN']<10)]
df_minority = df[(df['ZN']>10)]
df_minority_upsampled = resample(df_minority,
                                replace = True,
                                n_samples = 350,
                                random_state = 42)
df = pd.concat([df_minority_upsampled, df_minority])
df.head()
```

# Finding the factors affecting the price of various region

```python
plt.subplots(figsize=(20,15))
sns.heatmap(df.corr(),cmap = "YlGnBu",annot=True)
plt.show()
```

```python
df = df.drop(["LSTAT"],axis =1)
```

```python
df['MEDV'].unique()
```

```python
df["MEDV"].value_counts()
```

```python
df["MEDV"].mean()
```

```python
from xgboost import XGBRegressor
from sklearn import metrics as me
from sklearn.model_selection import train_test_split
```

```python
x = df.drop("MEDV",axis=1)
y=df["MEDV"]
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=42,test_size=0.2)
```

```python
model=XGBRegressor()
model.fit(x_train,y_train)
```

```python
y_predt = model.predict(x_test)
```

```python
r2_score=me.r2_score(y_test,y_predt)
print(r2_score)
```

```python
mean_error = me.mean_absolute_error(y_test,y_predt)
print(mean_error)
```

```python
plt.scatter(y_test,y_predt)
plt.xlabel("actual price")
plt.ylabel("predicted price")
plt.title("Comparision")
plt.show()
```
